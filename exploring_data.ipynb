```python
threshold = 0.05 * len(df)

cols = df.columns[df.isna().sum() < threshold]

df = df.dropna(subset=cols)
```

```python
len(df.columns)
```

```python
df.isna().mean()*100
```

```python
df = df.dropna(thresh=len(df) * .5, axis=1)
```

```python
df.columns
```

```python
# num_df = df.select_dtypes(include='number')
# clean_df = num_df.dropna()
# sns.pairplot(clean_df)

# plt.show()
```

```python
normal_columns=['HomeElo', 'AwayElo', 'Form3Home','Form5Home','Form3Away','Form5Away']
for col in normal_columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.isna().mean()*100
```

```python
numerical_cols=df.select_dtypes(include=np.number)
categorical_cols=df.select_dtypes(include='string')
for col in numerical_cols.columns:
    df[col] = df[col].fillna(df[col].median())

# Fill non-numerical columns with "Unknown"
for col in categorical_cols.columns:
    df[col] = df[col].fillna("Unknown")
```

```python
elo_df=pd.read_csv('EloRatings.csv')
```

```python
elo_df.info()
```

```python
print("Top 10 Clubs by Elo Rating:")
top_clubs = elo_df.sort_values(by='elo', ascending=False).head(10)
print(top_clubs)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_clubs['club'], y=top_clubs['elo'], hue=top_clubs['country'])
plt.title('Top 10 Clubs by Elo Rating')
plt.xlabel('Club')
plt.ylabel('Elo Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
print("Lowest 10 Clubs by Elo Rating:")
bottom_clubs = elo_df.sort_values(by='elo', ascending=True).head(10)
print(bottom_clubs)
plt.figure(figsize=(10, 6))
sns.barplot(x=bottom_clubs['club'], y=bottom_clubs['elo'], hue=bottom_clubs['country'])
plt.title('Lowest 10 Clubs by Elo Rating')
plt.xlabel('Club')
plt.ylabel('Elo Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
average_elo_by_country = elo_df.groupby('country')['elo'].mean().sort_values(ascending=False)
print("\nTop 10 Countries by Average Elo Rating:")
print(average_elo_by_country.head(10))

plt.figure(figsize=(10, 6))
sns.barplot(x=average_elo_by_country.head(10).index, y=average_elo_by_country.head(10).values)
plt.title('Top 10 Countries by Average Elo Rating')
plt.xlabel('Country')
plt.ylabel('Average Elo Rating')
plt.xticks(rotation=45)
plt.show()
```
